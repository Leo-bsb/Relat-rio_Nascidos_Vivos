# Gráficos

```python
diferentes = df[df['grupo_apgar1'] != df['grupo_apgar5']]
transicoes = diferentes.groupby(['grupo_apgar1', 'grupo_apgar5']).size().reset_index(name='Quantidade')
transicoes = transicoes[transicoes['Quantidade'] != 0]
total_transicoes = transicoes['Quantidade'].sum()
transicoes['Porcentagem'] = transicoes['Quantidade'] / total_transicoes * 100

ordem = ['Asfixia Grave → Asfixia Moderada',
         'Asfixia Grave → Boa vitalidade',
         'Asfixia Moderada → Boa vitalidade',
         'Boa vitalidade → Asfixia Moderada',
         'Boa vitalidade → Asfixia Grave',
         'Asfixia Moderada → Asfixia Grave']

cores = ['green', 'green', 'green', 'red', 'red', 'red']

transicoes['ordem'] = transicoes['grupo_apgar1'].astype(str) + ' → ' + transicoes['grupo_apgar5'].astype(str)
transicoes['ordem_num'] = transicoes['ordem'].apply(lambda x: ordem.index(x) if x in ordem else len(ordem))
transicoes = transicoes.sort_values('ordem_num')

plt.figure(figsize=(12, 7))
for index, row in transicoes.iterrows():
    cor = cores[ordem.index(row['ordem'])] if row['ordem'] in ordem else 'gray'
    label = f"{row['Quantidade']} ({row['Porcentagem']:.1f}%)"
    plt.bar(f"{row['grupo_apgar1']} → {row['grupo_apgar5']}", row['Quantidade'], color=cor)
    plt.text(f"{row['grupo_apgar1']} → {row['grupo_apgar5']}",
             row['Quantidade'] + (total_transicoes * 0.02),
             label,
             ha='center',
             fontsize=10,
             color='black')

plt.xticks(rotation=45, ha='right')
plt.xlabel('Transições')
plt.ylabel('Quantidade')
plt.title('Transições entre Grupos Apgar')
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import spearmanr


map_RaçaeCordoBebê = {
    'Preta ': 1,
    'Parda': 2,
    'Indígena': 3,
    'Branca': 4,
    'Amarela': 5
}

map_RaçaeCordaGestante = {
    'Preta ': 1,
    'Parda': 2,
    'Indígena': 3,
    'Branca': 4,
    'Amarela': 5
}

db = df.copy()
db['Raça e Cor do Bebê'] = db['Raça e Cor do Bebê'].map(map_RaçaeCordoBebê)
db['Raça e Cor da Gestante'] = db['Raça e Cor da Gestante'].map(map_RaçaeCordaGestante)

db = db.dropna(subset=['Raça e Cor do Bebê', 'Raça e Cor da Gestante'])

correlacao, p_valor = spearmanr(db['Raça e Cor do Bebê'], db['Raça e Cor da Gestante'])

print(f"Correlação de Spearman: {correlacao:.2f}")
print(f"p-valor: {p_valor:.4f}")
```
